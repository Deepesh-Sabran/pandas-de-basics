```python
import pandas as pd

netflix = pd.read_csv("netflix_titles.csv")
ratings = pd.read_csv("imdb_ratings.csv")

print(netflix)

print("-------------------")

print(ratings)
```

```python
## Difficulty Normal: (Merge)
# --> Which Netflix Movie has the highest IMDb Rating? [ Return only -> "title", "director", "country", "Rating" ]
merged_data = netflix.merge(
    ratings.drop_duplicates(subset=["title", "release_year"]),
    on=["title", "release_year"],
    how="left"
)

netflixMovieHighestIMDBRating = (
    merged_data.loc[
        merged_data["type"] == "Movie",
        ["title", "director", "country", "Rating"]
    ]
    .sort_values(by=["Rating"], ascending=False)
    .head(1)
)
print(netflixMovieHighestIMDBRating.to_string(index=False))
```

```python
## Difficulty Medium:
# --> Top 10 countries having Movies with IMDb Rating > 8.
top10CountyWithMovieRatingMoreThan8 = (
    merged_data.loc[
        (merged_data["type"] == "Movie") &
        (merged_data["Rating"] > 8),
        ["country"]
    ]
    .value_counts()
    .head(10)
)
print(top10CountyWithMovieRatingMoreThan8)
```

```python
# --> Find directors having at least 3 Netflix titles.
# directorsAtLeast3NetflixTitles = merged_data.groupby("director")["title"].size()
directorsAtLeast3NetflixTitles = merged_data["director"].value_counts()
print(directorsAtLeast3NetflixTitles[directorsAtLeast3NetflixTitles > 2])
```

```python
# --> Find content released after 2015 having IMDb Rating above 8.
contentAfter2015AndRatingAbove8 = (
    merged_data.loc[
        (merged_data["release_year"] > 2015) &
        (merged_data["Rating"] > 8)
    ]
)
print(contentAfter2015AndRatingAbove8.to_string(index=False))
```

```python
# --> Find all Movies with: [ "Rating > 8", "country = India" ]
moviesRatingAbove8CountryIndia = (
    merged_data.loc[
        (merged_data["type"] == "Movie") &
        (merged_data["Rating"] > 8) &
        (merged_data["country"].str.contains("India", na=False))
    ]
)
print(moviesRatingAbove8CountryIndia.to_string(index=False))
```

```python
## Difficulty Hard:
# --> Create a report: | Director | Movies | Avg IMDb Rating |, {{ sort by: Average Rating DESC, Movies DESC }}
report = (
    merged_data.loc[
        merged_data["type"] == "Movie"        # Filter based on type
    ]
    .groupby("director")["Rating"]            # grouping based on director and accessing Rating only
    .agg(
        Movies="count",
        Avg_IMDB_Rating="mean"
    )                                         # Movies & Avg_IMDB_Rating is a custom column name -> ["value passes to these variables is basically the instruction to pandas"]
    .reset_index()
).sort_values(
    by=["Avg_IMDB_Rating", "Movies"],
    ascending=[False, False]
)

print(report.to_string(index=False))
```
