```python
import pandas as pd

netflix = pd.read_csv("netflix_titles.csv")
ratings = pd.read_csv("imdb_ratings.csv")

print(netflix)
print("-------------------------")
print(ratings)
```

```python
# --> Merge both datasets on: "title", "release_year" keeps only matching records. Expected col: "title", "release_year", "type", "Rating", "director", "country"
merged_data = netflix.merge(ratings, on=["title", "release_year"], how="inner")[["title", "release_year", "type", "Rating", "director", "country"]]
print(merged_data)
```

```python
# --> After merging, ( Print: "title", "Rating", "type" )
print(merged_data[["title", "Rating", "type"]])
```

```python
# --> Find the Top 20 IMDb-rated Netflix titles.  ( Expected: "title", "Rating" sorted descending. )
top_20_imdbRatedNetflixTitle = merged_data.sort_values(by="Rating", ascending=False).head(20)[["title", "Rating"]]
print(top_20_imdbRatedNetflixTitle)
```

```python
# --> Find the average IMDb Rating of: "Movies", "TV Shows"
avgRating = merged_data.groupby("type")["Rating"].mean()
print(avgRating)
```

```python
# --> Find the highest-rated Movie.
highestRatedMovie = (
    merged_data.loc[
        merged_data["type"] == "Movie", ["title", "Rating"]
    ]
    .sort_values(by="Rating", ascending=False)
    .head(1)
)
print(highestRatedMovie["title"].values[0])
```

```python
# --> Find the highest-rated TV Show.
highestRatedTVShow = (
    merged_data.loc[
        merged_data["type"] == "TV Show",
        ["title", "Rating"]
    ]
    .sort_values(by="Rating", ascending=False)
    .head(1)
)
print(highestRatedTVShow["title"].values[0])
```

```python
# --> Find the Top 10 directors based on average IMDb Rating.
top10DirectorOnAvgIMDBRating = (
    merged_data.groupby("director")["Rating"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)
print(top10DirectorOnAvgIMDBRating)
```

```python
# --> Find Top 10 countries by average IMDb Rating.
top10CountriesOnAvgIMDBRating = (
    merged_data.groupby("country")["Rating"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)
print(top10CountriesOnAvgIMDBRating)
```

```python
# --> Find which release year has the highest average IMDb Rating.
releaseYrHighestAvgIMDBRating = merged_data.groupby("release_year")["Rating"].mean().sort_values(ascending=False).head(1)
print(releaseYrHighestAvgIMDBRating)
```

```python
# --> Find the oldest title whose IMDb Rating is above 8.
oldestTitleIMDBRatingOver8 = (
    merged_data.loc[
        merged_data["Rating"] > 8,
        ["title", "release_year"]
    ]
    .sort_values(by="release_year")
    .head(1)
)
print(oldestTitleIMDBRatingOver8["title"].values[0])
```

```python
## Left Join
left_merge = netflix.merge(ratings, on=["title", "release_year"], how="left")

# --> How many Netflix titles don't have an IMDb rating?
ratingNull = left_merge["Rating"].isnull().sum()
print(ratingNull)
```

```python
# --> List all Netflix titles missing IMDb ratings. [ Expected: "title", "release_year" ]
titleMissingRating = left_merge.loc[left_merge["Rating"].isnull(), ["title", "release_year"]]
print(titleMissingRating)
```

```python
## Right Join
right_merge = netflix.merge(ratings, on=["title", "release_year"], how="right", indicator=True)

# --> Are there IMDb titles that are not available on Netflix?
titleMissingOnNetflix = right_merge.loc[right_merge["show_id"].isnull(), ["title"]]
print(titleMissingOnNetflix)
```

```python
## Outer Join
outer_merge = netflix.merge(ratings, on=["title", "release_year"], how="outer", indicator=True)

# --> Find Total matched titles
totalMatchedTitles = outer_merge.loc[
    (outer_merge["show_id"].notnull()) &
    (outer_merge["Rating"].notnull())
].shape[0]
print(totalMatchedTitles)
```

```python
# --> Netflix-only titles
netflixTitles = outer_merge.loc[
    (outer_merge["show_id"].notnull()) &
    (outer_merge["Rating"].isnull()),
    ["title"]
]
print(netflixTitles)
```

```python
# --> IMDb-only titles
imdbTitles = outer_merge.loc[
    (outer_merge["show_id"].isnull()) &
    (outer_merge["Rating"].notnull()),
    ["title"]
]
print(imdbTitles)
```
