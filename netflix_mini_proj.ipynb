```python
import pandas as pd

netflix = pd.read_csv("netflix_titles.csv", index_col=0)
ratings = pd.read_csv("imdb_ratings.csv", index_col=0)

print(netflix)

print("______________________________")

print(ratings)
```

```python
## Level 1

# --> Total Movies vs TV Shows.
total = netflix.groupby("type").size()
# total = netflix["type"].value_counts()
print(total)

print("______________________________")

# --> Latest release year.
latestReleaseYr = netflix["release_year"].max()
print(latestReleaseYr)

print("______________________________")

# --> Oldest release year.
oldestReleaseYr = netflix["release_year"].min()
print(oldestReleaseYr)

print("______________________________")

# --> Number of titles from India.
# titlesIndia = netflix[netflix["country"].str.contains("India", na=False)].shape[0]
titlesIndia = netflix[netflix["country"].str.contains("India", na=False)].count()["title"]
print(titlesIndia)

print("______________________________")

# --> Top 10 countries.
top10Countries = netflix["country"].value_counts().head(10)
print(top10Countries)

```

```python
## Level 2

# --> Top 10 directors by number of titles.
top10Directors = netflix["director"].value_counts().head(10)
print(top10Directors)

print("______________________________")

# --> Count Movies and TV Shows per country.
# totalMovieTv = netflix.groupby(["country", "type"]).size()
totalMovieTv = netflix[["country", "type"]].value_counts()          # It's better over selected columns
print(totalMovieTv)

print("______________________________")

# --> Find all content released after 2018.
contentAfter2018 = netflix[netflix["release_year"] > 2018]
print(contentAfter2018)

print("______________________________")

# --> Find titles with missing director.
titleWithMissingDirector = netflix.loc[netflix["director"].isnull()][["title"]]
print(titleWithMissingDirector)

print("______________________________")

# --> Remove duplicate titles.
removeDuplicateTitles = netflix["title"].drop_duplicates()
print(removeDuplicateTitles)

```

```python
## Level 3

# --> Top 5 countries with most Movies.
mostMoviesCountry = (
    netflix.loc[
        (netflix["type"] == "Movie"),
        ["country"]
    ]
    .value_counts()
    .head()
)
print(mostMoviesCountry)

print("______________________________")

# --> Top 5 countries with most TV Shows.
mostTVShowCountry = (
    netflix.loc[
        (netflix["type"] == "TV Show"),
        ["country"]
    ]
    .value_counts()
    .head()
)
print(mostTVShowCountry)

print("______________________________")

# --> For each year, count Movies released.
moviesEachYr = (
    netflix.loc[
        (netflix["type"] == "Movie"),
        ["release_year"]
    ]
    .value_counts()
)
print(moviesEachYr)

print("______________________________")

# --> Find director with highest number of Movies.
directorWithHighestNumMovies = (
    netflix.loc[
        netflix["type"] == "Movie",
        ["director"]
    ]
    .value_counts()
    .head(1)
)
# print(directorWithHighestNumMovies.index[0][0])     # to only show name no count
print(directorWithHighestNumMovies)

print("______________________________")

# --> Find country producing content across the most years.
data = netflix.groupby("country")["release_year"].nunique().sort_values(ascending=False).head(1)
print(data)
```

```python
## Level 4

# --> Top 10 countries by content
top10CountriesByContent = netflix["country"].value_counts().head(10)
print(top10CountriesByContent)

print("______________________________")

# --> Movies vs TV Shows trend by year
trend = netflix.groupby(["release_year", "type"]).size().sort_index()
print(trend)

# Note: It is called a "trend" because arranging data over a sequential timeline allows you to see patterns over time.

print("______________________________")

# --> Most common ratings
mostCommonRating = netflix["rating"].value_counts().head(1)
print(mostCommonRating)

print("______________________________")

# --> Top directors
topDirectors = netflix[['director']].value_counts()
print(topDirectors)

print("______________________________")

# --> Missing data analysis
missingDataAnalysis = netflix.isnull().sum()
print(missingDataAnalysis)

print("______________________________")

# --> Content growth over time
netflix["date_added"] = netflix["date_added"].str.strip().str.split(", ").str[-1]
contentGrowth = netflix.groupby("date_added").size().sort_index().cumsum()
print(contentGrowth)
```
