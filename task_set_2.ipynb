```python
import pandas as pd

netflix = pd.read_csv("netflix_titles.csv")
ratings = pd.read_csv("imdb_ratings.csv")

print(netflix)

print("----------------------------")

print(ratings)
```

```python
# --> Number of Rows and Columns

print(netflix.shape)

print("----------------------------")

print(ratings.shape)
```

```python
# --> Print all column names.

netflix_col = netflix.columns
print(netflix_col)

print("--------------------")

ratings_col = ratings.columns
print(ratings_col)
```

```python
# --> Find datatype of every column.
netflix_col_datatypes = netflix.dtypes
print(netflix_col_datatypes)

print("----------------------------------")

rating_col_datatypes = ratings.dtypes
print(rating_col_datatypes)
```

```python
# --> Which columns contain null values?
netflix_null = netflix.columns[netflix.isnull().any()]
print(netflix_null)
```

```python
# --> How many unique ratings exist?
unique_ratings = netflix["rating"].nunique()
print(unique_ratings)
```

```python
# --> Get first row.
first_row = netflix.iloc[0]
print(first_row)
```

```python
# --> Get first 10 rows.
first_10_rows = netflix.iloc[:10]
print(first_10_rows)
```

```python
# --> Get rows 20-30 and only title, country, release_year:
rows_20_30 = netflix.iloc[20:30],[["title", "country", "release_year"]]
print(rows_20_30)
```

```python
# --> Get last 5 rows using iloc.
last_5_rows = netflix.iloc[-5:]
print(last_5_rows)
```

```python
# --> Get first 3 rows and first 4 columns.
rows_3_col_4 = netflix.iloc[:3, :4]
print(rows_3_col_4)
```

```python
# --> Find newest content.
newest_content = netflix.sort_values(
    by="date_added",
    key=lambda x: pd.to_datetime(x.str.strip()),    # lambda is no name function, you need to pass (lambda input: output)
    ascending=False
)

# newest_content = netflix.sort_values(by="release_year", ascending=False)
print(newest_content)
```

```python
# --> Find oldest content.
oldest_content = netflix.sort_values(
    by="date_added",
    key=lambda x: pd.to_datetime(x.str.strip()),
)

print(oldest_content)
```

```python
# --> Find top 20 latest TV Shows.
top_20_tv_shows = (
    netflix.loc[
        netflix["type"] == "TV Show",
        ["title", "release_year"]
    ]
    .sort_values(by="release_year")
    .head(20)
)

print(top_20_tv_shows)
```

```python
# --> sort by country ASC, release_year DESC
sort = netflix.sort_values(by=["country", "release_year"], ascending=[True, False])
print(sort)
```
